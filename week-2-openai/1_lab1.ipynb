```python
# The imports

from dotenv import load_dotenv
from agents import Agent, Runner, trace, OpenAIChatCompletionsModel, set_tracing_export_api_key
from openai import AsyncOpenAI
import os
```

```python
# The usual starting point

load_dotenv(override=True)
# This key is ONLY used to upload the visual graph blocks, not to run models
set_tracing_export_api_key(os.getenv("OPENAI_API_KEY"))
```

```python
DEEPSEEK_BASE_URL = "https://api.deepseek.com/v1"

deepseek_client = AsyncOpenAI(base_url=DEEPSEEK_BASE_URL, api_key=os.getenv("DEEPSEEK_API_KEY"))

deepseek_model = OpenAIChatCompletionsModel(model="deepseek-v4-flash", openai_client=deepseek_client)
```

```python

# Make an agent with name, instructions, model

agent = Agent(
    name="Jokester", 
    instructions="You are a joke teller. Tell brief, witty tech jokes.", 
    model=deepseek_model
)
```

```python
# Run the joke with Runner.run(agent, prompt) then print final_output

with trace("Telling a joke"):
    result = await Runner.run(agent, "Tell a joke about Autonomous AI Agents")
    print(result.final_output)
```
